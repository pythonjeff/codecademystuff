```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
bc_data = load_breast_cancer()
```

```python
bc_data.feature_names
```

```python
bc_data.target_names
```

```python
bc_data.target
```

```python
training_data, validation_data, training_labels, validation_labels = train_test_split(bc_data.data, bc_data.target, test_size=0.2, random_state=100)
```

```python
print(len(training_data))
print(len(training_labels))
```

```python
accuracies = []
k_list = range(1, 101)

for k in range(1, 101):
    
    classifier = KNeighborsClassifier(n_neighbors=k)
    classifier.fit(training_data, training_labels)
    accuracy = classifier.score(validation_data, validation_labels)
    accuracies.append(accuracy)
```

```python
plt.plot(k_list, accuracies)
plt.xlabel('k')
plt.ylabel('validation accuracy')
plt.title('Breast Cancer Classifier Accuracy')
plt.show()
```
